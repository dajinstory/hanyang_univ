```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
for func in ['Gaussian Distribution', 'Uniform Distribution']:
    for samples in ['100','10000','1000000']:
        for sections in ['50', '100']:
            filename = func+'_'+samples+'_'+sections
            df=pd.read_csv('distributions/'+filename+'.csv')

            plt.rcParams["figure.figsize"] = (20,10)
            plt.title(filename)
            plt.xlabel('random_number')
            plt.ylabel('frequency')
            plt.xlim(-6, 6)
            plt.ylim(0, df['cnt'].max()*1.5)
            index = np.arange(int(sections))
            plt.xticks(rotation=90)

            plt.bar(index, df['cnt'], tick_label=df['x'])
            plt.savefig(filename)
            plt.show()
            plt.close()
```
